```python
#Alexandria Benedict, Assignment 5


from random import random
import nltk
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
import pandas as pd
from math import log
from sklearn.model_selection import train_test_split
from collections import *
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
import numpy as np
from sklearn import metrics
```

## Question 1

```python

#Take in data
df1 = pd.read_csv('CONcreTEXT_trial_EN.tsv', sep='\t', header=None, engine= 'python')
df2 = pd.read_csv('CONcreTEXT_trial_IT.tsv', sep='\t', header=None, engine= 'python')


df1.at[0, 5] = "LANGUAGE"
df1[5][1:] = "English"
df2.at[0, 5] = "LANGUAGE"
df2[5][1:] = "Italian"
#df2.head()

df1.at[0, 6] = "CONCRETE"

row = 1
for i in df1[4][1:]:
    if float(df1[4][row]) <= 4:
        df1[6][row] = 'LOW'
    else:
        df1[6][row] = 'HIGH'
    #df1[6][row] = np.where(float(df1[4][row]) <= 4, 'LOW', 'HIGH')
    row += 1

df2.at[0, 6] = "CONCRETE"

row2 = 1
for i in df2[4][1:]:
    if float(df2[4][row2]) <= 4:
        df2[6][row2] = 'LOW'
    else:
        df2[6][row2] = 'HIGH'
    #df2[6][row2] = np.where(float(df2[4][row2]) <= 4, 'LOW', 'HIGH')
    row2 += 1



df = pd.concat([df1,df2])
df.at[0, 7] = "PREDICTIONS"
#print(df)
#print(len(df))
df.head()
```

## Question 2

```python
train, test = train_test_split(df, test_size=0.2, train_size = 0.8, random_state=42)
```

## Question 3

```python
def majorityClass(train):
    row = 1
    high = 0
    low = 0
    for i in train[6][1:]:
        #print(i)
        if i == "HIGH":
            high += 1
        else:
            low += 1
        row += 1 
    #print("High: ", high)
    #print("Low: ", low)
    if high > low: 
        return "HIGH"
    else:
        return "LOW"
    

test[7][:] = majorityClass(train)

target_names = ["HIGH", "LOW"]
y_true = []
y_pred = []

for i in range(len(test)):
    y_pred.append(test[7].iloc[i])
    y_true.append(test[6].iloc[i])
#test.head()

print(metrics.classification_report(y_true, y_pred, target_names))

print(metrics.confusion_matrix(y_true, y_pred))
```

## Question 4

```python
test.head()
row = 0
for i in test[0][1:]:
    #print(i)
    test[7].iloc[row] = np.where(len(test[0].iloc[row]) >= 5, 'HIGH', 'LOW')
    row += 1
test.head()


target_names = ["HIGH", "LOW"]
y_true = []
y_pred = []

for i in range(len(test)):
    y_pred.append(test[7].iloc[i])
    y_true.append(test[6].iloc[i])
#test.head()

print(metrics.classification_report(y_true, y_pred, target_names))

print(metrics.confusion_matrix(y_true, y_pred))
```

## Question 5

```python
#Text preprocessing
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(df[3][:])
#X_train_counts.shape



tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape

nb_train = []

for i in range(len(test)):
    #print(test[6].iloc[i])
    if test[6].iloc[i] == "HIGH":
        nb_train.append("HIGH")
    else:
        nb_train.append("LOW")
#print(nb_train)
#Train Naive Bayes Classifier
clf = MultinomialNB().fit(X_train_tfidf, df[6][:])

docs_new = []

for i in range(len(test)):
    docs_new.append(test[3].iloc[i])

#print(docs_new)


X_new_counts = count_vect.transform(docs_new)
X_new_tfidf = tfidf_transformer.transform(X_new_counts)

predicted = clf.predict(X_new_tfidf)
y_pred = []
for doc, category in zip(docs_new, predicted):
    print('%r => %s' % (doc, category))
    y_pred.append(category)

    
    
print(metrics.classification_report(y_true, y_pred, target_names))

print(metrics.confusion_matrix(y_true, y_pred))
```

```python
test.head()
def thresholdClass(threshold):
    row = 0
    for i in test[0][1:]:
        #print(i)
        test[7].iloc[row] = np.where(len(test[0].iloc[row]) >= threshold, 'HIGH', 'LOW')
        row += 1
def getResults():
    y_true = []
    y_pred = []

    for i in range(len(test)):
        y_pred.append(test[7].iloc[i])
        y_true.append(test[6].iloc[i])

    print(metrics.classification_report(y_true, y_pred, target_names))
    print(metrics.confusion_matrix(y_true, y_pred))

    
thresholdClass(2)
getResults()

    
thresholdClass(5)
getResults()


thresholdClass(6)
getResults()
```

I chose the thresholds above just to experiment values different from Question 4. Using a much smaller threshold such as 2 shows that there is very little to no classification to "LOW" as most words would be larger than 2 characters. As the threshold increases, you can see that there is more classifcation skewed to LOW rather than HIGH, as classifying words to the HIGH category would become more difficult with a higher threshold.
